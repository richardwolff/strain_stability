```python
import matplotlib.pyplot as plt
import miscellaneous_utils as mu
import pandas as pd
import numpy as np
```

```python
import matplotlib

matplotlib.rc('xtick', labelsize=50) 
matplotlib.rc('ytick', labelsize=50) 

plt.rcParams["xtick.major.size"] = 0
plt.rcParams["ytick.major.size"] = 0


plt.rcParams.update({
  "text.usetex": True,
  "font.family": "Sans-serif"
})
```

```python
palette = ["#a3606f","#ffcc5c","#6769a9","#4aab80"]
```

```python
dates_am = mu.return_dates("am")
dates_ao = mu.return_dates("ao")
dates_an = mu.return_dates("an")
dates_ae = mu.return_dates("ae")
```

```python
fig,ax = plt.subplots(figsize=(20,16))

ax.scatter(dates_am.values,2*np.ones(len(dates_am.values)),s=200,edgecolor="k",alpha=.8,color=palette[0])
#ax.plot(dates_am.values,np.ones(len(dates_am.values)))

ax.scatter(dates_ao.values,3*np.ones(len(dates_ao.values)),s=200,edgecolor="k",alpha=.8,color=palette[1])
ax.scatter(dates_an.values,4*np.ones(len(dates_an.values)),s=200,edgecolor="k",alpha=.8,color=palette[2])
ax.scatter(dates_ae.values,5*np.ones(len(dates_ae.values)),s=200,edgecolor="k",alpha=.8,color=palette[3])

labels = ["am","ao","an","ae"]

ax.set_yticks([2,3,4,5])
ax.set_yticklabels(labels,va='center')

ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["left"].set_visible(False)

#ax.spines["left"].set_linewidth(6)
ax.spines["bottom"].set_linewidth(0)

ax.set_ylim([1.75,5.25])

ax.set_xlabel("Time (days)",size=50,labelpad=30)

#ax.grid(True,alpha=1)
```

```python
from matplotlib.ticker import MaxNLocator
```

```python
matplotlib.rc('xtick', labelsize=50) 
matplotlib.rc('ytick', labelsize=25) 

bins = np.arange(0,75,5)

fig,axs = plt.subplots(4,1,figsize=(6,16))

for a in axs:
    a.set_xlim([0,70])
    a.set_xticks([0,25,50])
    a.yaxis.tick_right()
    a.yaxis.set_ticks_position('both')
    a.yaxis.set_major_locator(MaxNLocator(4)) 
    a.spines["top"].set_visible(False)
    a.spines["left"].set_visible(False)
    
axs[0].hist(dates_ae.diff()[1:].values, edgecolor="k",color=palette[3],bins=bins,zorder=10)
#axs[0].semilogy()
#axs[0].set_xlim([0,60])

axs[1].hist(dates_an.diff()[1:].values, edgecolor="k",color=palette[2],bins=bins,zorder=10)
#axs[1].semilogy()
#axs[1].set_xlim([0,60])

axs[2].hist(dates_ao.diff()[1:].values, edgecolor="k",color=palette[1],bins=bins,zorder=10)
#axs[2].semilogy()
#axs[2].set_xlim([0,60])

axs[3].hist(dates_am.diff()[1:].values, edgecolor="k",color=palette[0],bins=bins,zorder=10)
#axs[3].semilogy()
#axs[3].set_xlim([0,60])

axs[3].set_xlabel(r"$\Delta$ Time (days)",size=50,labelpad=30)

fig.subplots_adjust(hspace=1)
```
