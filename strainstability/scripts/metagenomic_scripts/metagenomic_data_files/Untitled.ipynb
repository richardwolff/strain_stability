```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
from matplotlib.ticker import MaxNLocator

matplotlib.rc('xtick', labelsize=50) 
matplotlib.rc('ytick', labelsize=50) 

plt.rcParams["xtick.major.size"] = 8
plt.rcParams["xtick.major.width"] = 3

plt.rcParams["xtick.minor.size"] = 6
plt.rcParams["xtick.minor.width"] = 2

plt.rcParams["ytick.major.size"] = 0


plt.rcParams.update({
  "text.usetex": True,
  "font.family": "Sans-serif"
})

palette = ["#a3606f","#ffcc5c","#6769a9","#4aab80"]
palette_dic = {"am":"#a3606f","ao":"#ffcc5c","an":"#6769a9","ae":"#4aab80"}
```

```python
df = pd.read_csv("Poyet_sample_metadata.csv",sep="\t",index_col=0)
```

```python
df_gb = df.groupby("host")
```

```python
fig,axs = plt.subplots(1,4,figsize=(60,12))

i = 0
for host in ["am","ao","an","ae"]:
    axs[i].set_title(host,size=80)
    axs[i].hist(df_gb.get_group(host)["read_count"].values,edgecolor="k",color=palette_dic[host])
    axs[i].yaxis.set_major_locator(MaxNLocator(4)) 
    i+=1
    
axs[0].set_ylabel("\# Samples",size=80,labelpad=40)

fig.text(.475,-.1,"Read count",size=80)
```
