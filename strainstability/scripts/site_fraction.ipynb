```python
import pandas as pd
import matplotlib.pyplot as plt
import config
```

```python
data_dir = "/u/scratch/r/rwolff/Poyet_midas_output_spu/merged_midas_output/%s/snps"

X = {"am":{},"ao":{},"an":{},"ae":{}}

for host in ["ao","am","an","ae"]:
    
    snps_dir = data_dir % host
    species_list = pd.read_csv("%s/species_snps.txt" % snps_dir,header=None)[0].values
    
    for species in species_list:
        snps_spec_dir = ("%s/%s" % (snps_dir,species))
        X[host][species] = return_num_good_sites(snps_spec_dir)
        
```

```python
def return_num_good_sites(snps_spec_dir):
    
    min_depth_pi_site = config.min_depth_pi_site
    
    snps_summary = pd.read_csv("%s/snps_summary.txt" % snps_spec_dir,sep="\t",index_col=0)
    L = snps_summary["covered_bases"]
   
    chunk_size = 40000
    samples_host = list(pd.read_csv("%s/snps_depth.txt.bz2" % snps_spec_dir,sep="\t",index_col=0, nrows=0))
    
    df_depth_reader = pd.read_csv("%s/snps_depth.txt.bz2" % snps_spec_dir,sep="\t",index_col=0, iterator=True,low_memory=False)
    
    df_depth = df_depth_reader.get_chunk(0)
    
    df_pass_prop = pd.Series(0,index=samples_host)
    df_tot_length = 0
    
    reader=True
    
    while reader:

        df_depth = df_depth_reader.get_chunk(chunk_size)
        
        df_pass_prop += (df_depth > 5).sum()
        df_tot_length += df_depth.shape[0]
        
        if df_depth.shape[0] < chunk_size:
            reader=False
            print("Complete")
            
    print(df_pass_prop/df_tot_length)        
    return df_pass_prop/df_tot_length
```
